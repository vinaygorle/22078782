```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import textwrap  # Import textwrap for wrapping the abstract text

# Reading data set from the local folder
data = pd.read_csv('world_nuclear_power_dataset.csv')

# Using tail to take the last 50 data values
data = data.tail(50)

# Create a figure and subplots with smaller sizes
fig, axs = plt.subplots(2, 2, figsize=(12, 10))

# Set background color
fig.patch.set_facecolor('tomato')
fig.suptitle('Nuclear Power Plant Analysis', fontsize=16, color='white')  # Adjusted title color

# Bar plot about the status of the power plant
sns.countplot(x='Status', data=data, ax=axs[1, 1], palette='dark:peru')
axs[1, 1].set_title('Status of the Power Plant')

# Pie plot for type of fuel used 
status_counts = data['Fuel'].value_counts()
axs[0, 1].pie(status_counts, labels=status_counts.index, autopct='%1.1f%%', startangle=140, colors=sns.color_palette("Pastel1"))
axs[0, 1].legend(bbox_to_anchor=(1, 1), loc='upper left', labels=status_counts.index)
axs[0, 1].set_title('Type of fuel used in nuclear power plant')

# Visualization 3: Histogram plot
sns.histplot(data['Capacity_MW'], bins=20, kde=True, ax=axs[1, 0], color='orangered')
axs[1, 0].set_title('Total power plant capacity')
axs[1, 0].set_xlabel('Capacity (MW)')
axs[1, 0].set_ylabel('Frequency')

# Line plot instead of scatter plot
sns.lineplot(x=data['Year_Commissioned'], y=data['Construction_Start'], ax=axs[0, 0])  # Changed to line plot
axs[0, 0].set_title('Year of commissioned vs construction start')
axs[0, 0].set_xlabel('Year of Commissioned for nuclear power plant')
axs[0, 0].set_ylabel('Construction Start date')

# Stylish box for author information
author_box = dict(boxstyle='round', facecolor='#FFD700', alpha=0.5)
fig.text(0.08, 0.01, "Name: Gorle venkata vinay prem chand", ha="left", va="bottom", fontsize=10, color='black', bbox=author_box)
fig.text(0.08, 0.05, "Student ID: 22078782", ha="left", va="bottom", fontsize=10, color='black', bbox=author_box)
# Replace abstract text based on plot observations
abstract_text = (
    "The highest number of plants started construction in the 1970s and were commissioned in the 1980s. "
    "There has been a recent increase in construction starts and commissioning, with a few plants starting construction "
    "in the 2010s and being commissioned in the 2020s. "
    "Uranium is the most common fuel type (32%), followed by mixed oxide fuel (30%) and thorium (30%). "
    "The majority of nuclear power plants are operational (38%), followed by decommissioned (32%), under construction (14%), "
    "and planned (16%). "
    "Total power plant capacity is highest for plants that were commissioned in the 1980s, with a capacity of over 300,000 MW seen in histogram."
)

# Wrap the abstract text for better readability
wrapped_abstract_text = textwrap.fill(abstract_text, width=50)

# Adjusted text position to avoid overlap
fig.text(0.4, 0.06, textwrap.fill(abstract_text , width=120), horizontalalignment='center', fontsize=10, color='black', wrap=True)  # Adjusted text color
plt.subplots_adjust(top=0.9, right=0.85, bottom=0.2)  # Adjust space to fit everything
# Save the dashboard as an image
plt.savefig('22078782.png', dpi=300)  # Change the file format
```
